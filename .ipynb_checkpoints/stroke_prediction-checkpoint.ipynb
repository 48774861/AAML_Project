```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
orig_data = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
pp_data = orig_data.drop("id", axis=1)
pp_data = pp_data.dropna(subset=['bmi'])
pp_data.reset_index(inplace=True, drop=True)

le = LabelEncoder()
le_cols = ['gender', 'ever_married', 'Residence_type']
for col in le_cols:
    pp_data[col] = le.fit_transform(pp_data[col])

ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_cols = ['work_type', 'smoking_status']
encoded = pd.DataFrame(ohe.fit_transform(pp_data[ohe_cols]), columns=ohe.get_feature_names_out(ohe_cols))
encoded = encoded.drop('smoking_status_Unknown', axis=1)
pp_data = pd.concat([pp_data.drop(columns=ohe_cols), encoded], axis=1)

pp_data
```
