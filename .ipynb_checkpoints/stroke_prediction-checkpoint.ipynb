```python
def oversample(df):
    """
    Oversamples the minority class in the 'stroke' column of a DataFrame.

    Parameters:
    df (pd.DataFrame): Input DataFrame with a binary 'stroke' column.

    Returns:
    pd.DataFrame: A new DataFrame with oversampled minority class.
    """
    # Separate majority and minority classes
    df_majority = df[df['stroke'] == 0]
    df_minority = df[df['stroke'] == 1]

    # Oversample minority class
    df_minority_oversampled = resample(
        df_minority,
        replace=True,                      # Sample with replacement
        n_samples=len(df_majority),        # Match number of majority class
        random_state=42                    # For reproducibility
    )

    # Combine majority and oversampled minority
    df_oversampled = pd.concat([df_majority, df_minority_oversampled])

    # Shuffle the dataset
    df_oversampled = df_oversampled.sample(frac=1, random_state=42).reset_index(drop=True)

    return df_oversampled
```

```python
X = pp_data.drop("stroke", axis=1)
y = pp_data["stroke"]

pp_data_train, pp_data_test = train_test_split(
    pp_data, test_size=0.25, stratify=pp_data['stroke'], random_state=42
)
# pp_data_train = oversample(pp_data_train)

X_train = pp_data_train.drop("stroke", axis=1)
y_train = pp_data_train["stroke"]

sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

# pp_data_test = oversample(pp_data_test)
X_test = pp_data_test.drop("stroke", axis=1)
y_test = pp_data_test["stroke"]

# sm = SMOTE(random_state=42)
# X_test, y_test = sm.fit_resample(X_test, y_test)

# Convert to DMatrix format
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

```

```python
import itertools

def grid_search_xgb(dtrain, dtest, y_test):
    # Define parameter ranges
    eta_vals = [0.001, 0.01, 0.03, 0.05, 0.1, 0.3]
    max_depth_vals = [3, 5, 7]
    subsample_vals = [0.8, 0.6, 0.5]
    colsample_vals = [0.8, 0.6, 0.5]
    lambda_vals = [1, 5]
    alpha_vals = [0, 1]
    n_estimators_vals = [100, 300]
    objectives = ['binary:logistic']
    eval_metrics = ['auc']

    # Store results
    results = []

    count = 1
    print(f"Testing combination: ", end='')
    
    # Iterate through all combinations
    for eta, max_depth, subsample, colsample_bytree, reg_lambda, reg_alpha, n_estimators, objective, eval_metric in itertools.product(
        eta_vals, max_depth_vals, subsample_vals, colsample_vals,
        lambda_vals, alpha_vals, n_estimators_vals, objectives, eval_metrics
    ):
        params = {
            'eta': eta,
            'max_depth': max_depth,
            'subsample': subsample,
            'colsample_bytree': colsample_bytree,
            'lambda': reg_lambda,
            'alpha': reg_alpha,
            'objective': objective,
            'eval_metric': eval_metric
        }

        bst = xgb.train(params, dtrain, num_boost_round=n_estimators)

        # Predict and evaluate
        preds = bst.predict(dtest)
        y_pred = (preds > 0.5).astype(int)

        acc = accuracy_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)

        # Append results
        results.append({
            'eta': eta,
            'max_depth': max_depth,
            'subsample': subsample,
            'colsample_bytree': colsample_bytree,
            'lambda': reg_lambda,
            'alpha': reg_alpha,
            'n_estimators': n_estimators,
            'accuracy': acc,
            'auc': auc,
            'f1_score': f1,
            'precision': precision,
            'recall': recall
        })

        print(f"{count}, ", end='')
        count = count + 1

    # Convert to DataFrame and sort by F1 score
    results_df = pd.DataFrame(results)
    sorted_df = results_df.sort_values(by='f1_score', ascending=False).reset_index(drop=True)

    return sorted_df
```

We are actually going to use the Receiver Operating Characteristic Curve to examine how we can change the classification threshold for predicting strokes to impact the True Positive Rate and False Positive Rates. In essence, as the TPR increases, we may also increase the FPR as well (which is bad). Compared to the F1-score, the AUC under the ROC curve is more useful for imbalanced classes since it is independent of a specific threshold like being above or below 0.5 for example.

True Positive Rate is the % of predicted positives that are real (i.e. recall). False Positive Rate is the % of predicted positives that are actually negative.
We have around 5000 examples, where 5% have a stroke (positive instance). Thus, around 250 examples have a stroke, and 4750 don't.
With a TPR of 80%, with 250 actual positives, we get 0.8*250 = 200 of the people with strokes correctly predicted.
With a FPR of 0.26, with 4750 actual negatives, we get 0.26*4750 = 1235 people without strokes that we think we do. That's a lot of people we incorrectly identify.

Precision: Of all the predicted positives, how many were actually correct? Recall: Of all the actual positives, how many did we correctly identify? For XGB, we can have high recall (0.8) with low precision (0.2), which means we actually diagnose tons of people without a stroke. This isn't good.

Balanced testing dataset doesn't mean much for precision,recall, and F1 score since we are basically duplicating the same or fake data to predict the same again. It's balanced, but it's just with fake data that doesn't really mean much

1. Balancing the Test Set Skews Real-World Performance
Why it matters: In real-world scenarios, the data is often imbalanced — especially in medical datasets like stroke prediction.

What happens when you balance: You artificially inflate the number of minority class examples (e.g., stroke cases), which makes it easier for the model to detect them.

Impact on metrics:

Precision and recall may appear better than they would be in deployment.

F1 score becomes misleading because the class distribution no longer reflects reality.

AUC (especially ROC AUC) is more robust to imbalance, but still benefits from realistic class proportions.

🧪 2. Imputation Adds Another Layer of Uncertainty
Imputed values (whether by mean, MICE, or age-group methods) are not real observations.

They can smooth out signal or introduce bias, especially if the imputation method assumes patterns that don’t hold across all subgroups.

Comparing models on imputed data is fine for internal benchmarking, but external validation on raw or minimally imputed data is essential.

```python
import pandas as pd
import itertools
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

import pandas as pd
import itertools
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def grid_search_logreg(X_train, y_train, X_test, y_test):
    solvers = ['lbfgs', 'liblinear', 'saga']
    penalties = [None, 'l1', 'l2']
    Cs = [0.01, 0.1, 1.0, 10.0]
    max_iters = [100, 300, 500]

    results = []
    count = 1
    print("Testing combination: ", end='')

    for solver, penalty, C, max_iter in itertools.product(solvers, penalties, Cs, max_iters):
        # Skip incompatible combinations
        if penalty == 'l1' and solver not in ['liblinear', 'saga']:
            continue
        if penalty == None and solver not in ['lbfgs', 'saga']:
            continue

        try:
            model = LogisticRegression(solver=solver, penalty=penalty, C=C, max_iter=max_iter)
            model.fit(X_train, y_train)
            y_pred_prob = model.predict_proba(X_test)[:, 1]
            y_pred = (y_pred_prob > 0.5).astype(int)

            acc = accuracy_score(y_test, y_pred)
            auc = roc_auc_score(y_test, y_pred_prob)
            f1 = f1_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred)
            recall = recall_score(y_test, y_pred)

            results.append({
                'solver': solver,
                'penalty': penalty,
                'C': C,
                'max_iter': max_iter,
                'accuracy': acc,
                'auc': auc,
                'f1_score': f1,
                'precision': precision,
                'recall': recall
            })

            print(f"{count}, ", end='')
            count += 1

        except Exception as e:
            print(f"\nSkipped combination due to error: {e}")

    results_df = pd.DataFrame(results)
    sorted_df = results_df.sort_values(by='f1_score', ascending=False).reset_index(drop=True)
    return sorted_df
```
